TSNE helps visualizing data in 2D space.
In this example we use the digits dataset from sklearn.

```python
from sklearn.manifold import TSNE
from sklearn.datasets import load_digits
import pandas as pd
import matplotlib.pyplot as plt

digits = load_digits()
x = pd.DataFrame(digits.data)

model = TSNE()
x_transformed = model.fit_transform(x)
```

```python
plt.scatter(x_transformed[:,0], x_transformed[:,1])
plt.show()
```

As you can see, there are 9-10 clusters delimited. In the case that you need to apply KMeans and need to determine how many clusters will be required, this method can give you a good idea. The correct output in this case is 10, because the target variable contains the numbers from 0 to 9.
