TSNE helps visualizing data in 2D space.
In this example we use the digits dataset from sklearn.

```python
from sklearn.manifold import TSNE
from sklearn.datasets import load_digits
import pandas as pd
import matplotlib.pyplot as plt
```

```python
digits = load_digits()
x = pd.DataFrame(digits.data)
```

```python
model = TSNE()
x_transformed = model.fit_transform(x)
```

```python
plt.scatter(x_transformed[:,0], x_transformed[:,1], c=digits.target)
plt.show()
```

As you can see, there are 9-10 clusters delimited. I added colors to compare with the target data, but you should not expect to know that in advance when exploring this kind of data.
In the case that you need to apply KMeans and need to determine how many clusters will be required, this method can give you a good idea. The correct output in this case is 10, because the target variable contains the numbers from 0 to 9.
